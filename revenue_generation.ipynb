## Quarterly Retention Rate

```python
#identify which users are active by looking at their revenue per month
df_user_purchase_quarter = df.groupby(['customer_unique_id','order_purchase_quarter'])['total_money'].sum().reset_index()
df_user_purchase_quarter
```

```python
df_retention = pd.crosstab(df_user_purchase_quarter['customer_unique_id'], df_user_purchase_quarter['order_purchase_quarter'])

df_retention.head()
```

```python
#create an array of dictionary which keeps Retained & Total User count for each month
months = df_retention.columns[:]
retention_array = []
for i in range(len(months)-1):
    retention_data = {}
    selected_month = months[i+1]
    prev_month = months[i]
    #print(prev_month)
    retention_data['order_purchase_quarter'] = int(selected_month)
    retention_data['TotalUserCount'] = df_retention[selected_month].sum()
    retention_data['RetainedUserCount'] = df_retention[(df_retention[selected_month]>0) & (df_retention[prev_month]>0)][selected_month].sum()
    retention_array.append(retention_data)
    
#convert the array to dataframe and calculate Retention Rate
df_retention_new = pd.DataFrame(retention_array)
df_retention_new['RetentionRate'] = df_retention_new['RetainedUserCount']/df_retention_new['TotalUserCount']
df_retention_new
```
